### 2.Imputing Seed from TourneySeeds

```python

def impute_seeds(seeds_df,df):
    
    for i in range(len(seeds_df)):
        if len(seeds_df['Seed'][i]) ==3:
           seeds_df['Seed'][i] = seeds_df['Seed'][i][-2:]
        elif len(seeds_df['Seed'][i]) ==4:
             seeds_df['Seed'][i] = seeds_df['Seed'][i][1:3]
    
    df['WTeamSeed'] = df.merge(right = seeds_df.rename(columns = {'TeamID':'WTeamID'}),on=['Season','WTeamID'])['Seed']
    df['LTeamSeed'] = df.merge(right = seeds_df.rename(columns = {'TeamID':'LTeamID'}),on=['Season','LTeamID'])['Seed']
    return df
```

```python
TourneyDetResults = impute_seeds(TourneySeeds,TourneyDetResults)
TourneyCompResults = impute_seeds(TourneySeeds,TourneyCompResults)
```

```python


def preprocess_regular(dataframe):
    
    df = dataframe.drop(columns = ['NumOT','Season'])

    half1 = df.sample(n = int(len(df)*0.5-0.5))
    half2 = df[~df.isin(half1)].dropna()

    half1['WLoc'] = half1['WLoc'].replace({'A' : -1,'H': 1,'N':0})
    half2['WLoc'] = half2['WLoc'].replace({'A' : -1,'H': 0,'N':1})
    
    half1.columns.tolist()
    
    drop_columns = ['1Ast', '1Blk', '1DR', '1FGA', '1FGA3', '1FGM', '1FGM3', '1FTA', '1FTM','1OR', '1PF', '1Score', '1Stl', '1TO' , '2Ast', '2Blk', '2DR','2FGA', '2FGA3', '2FGM', '2FGM3', '2FTA', '2FTM', '2OR', '2PF','2Score', '2Stl', '2TO', 'DayNum']
    d1 = {'WLoc':'Team1_home','Season': 'Season','DayNum': 'DayNum','WTeamID': 'Team1','WScore': '1Score','LTeamID': 'Team2','LScore': '2Score','WFGM': '1FGM','WFGA': '1FGA','WFGM3': '1FGM3','WFGA3': '1FGA3','WFTM': '1FTM','WFTA': '1FTA','WOR': '1OR','WDR': '1DR','WAst': '1Ast','WTO': '1TO','WStl': '1Stl','WBlk': '1Blk','WPF': '1PF','LFGM': '2FGM','LFGA': '2FGA','LFGM3': '2FGM3','LFGA3': '2FGA3','LFTM': '2FTM','LFTA': '2FTA','LOR': '2OR','LDR': '2DR','LAst': '2Ast','LTO': '2TO','LStl': '2Stl','LBlk': '2Blk','LPF': '2PF','WTeamSeed': 'Team1Seed','LTeamSeed':'Team2Seed'}
    d2 = {'WLoc':'Team1_home','Season': 'Season','DayNum': 'DayNum','WTeamID': 'Team2','WScore': '2Score','LTeamID': 'Team1','LScore': '1Score','WFGM': '2FGM','WFGA': '2FGA','WFGM3': '2FGM3','WFGA3': '2FGA3','WFTM': '2FTM','WFTA': '2FTA','WOR': '2OR','WDR': '2DR','WAst': '2Ast','WTO': '2TO','WStl': '2Stl','WBlk': '2Blk','WPF': '2PF','LFGM': '1FGM','LFGA': '1FGA','LFGM3': '1FGM3','LFGA3': '1FGA3','LFTM': '1FTM','LFTA': '1FTA','LOR': '1OR','LDR': '1DR','LAst': '1Ast','LTO': '1TO','LStl': '1Stl','LBlk': '1Blk','LPF': '1PF','WTeamSeed': 'Team2Seed','LTeamSeed':'Team1Seed'}
    
    df = pd.concat([half1.rename(columns = d1),half2.rename(columns = d2)])
 
    df['FGM_diff'] = df['1FGM']- df['2FGM']
    df['FGA_diff'] = df['1FGA']- df['2FGA']
    df['FGA3_diff'] = df['1FGA3']- df['2FGA3']
    df['FGM3_diff'] = df['1FGM3']- df['2FGM3']
    df['FTA_diff'] = df['1FTM']- df['2FTM']
    df['FTA_diff'] = df['1FTA']- df['2FTA']
    df['OR_diff'] = df['1OR']- df['2OR']
    df['DR_diff'] = df['1DR']- df['2DR']
    df['Ast_diff'] = df['1Ast']- df['2Ast']
    df['TO_diff'] = df['1TO']- df['2TO']
    df['Stl_diff'] = df['1Stl']- df['2Stl']
    df['Blk_diff'] = df['1Blk']- df['2Blk']
    df['PF_diff'] = df['1PF']- df['2PF']
    df['score_diff'] = df['1Score']- df['2Score']
    
    
    df.drop(columns= drop_columns,inplace = True)

    return df
```

```python
def regular_compact_preprocessing(dataframe):
    df = dataframe.drop(columns = ['NumOT','Season','DayNum'])

    half1 = df.sample(n = int(len(df)*0.5-0.5))
    half2 = df[~df.isin(half1)].dropna()
    
    half1['WLoc'] = half1['WLoc'].replace({'A' : -1,'H': 1,'N':0})
    half2['WLoc'] = half2['WLoc'].replace({'A' : -1,'H': 0,'N':1})
    
    d1 = {'WTeamID': 'Team1','WScore': '1Score','LTeamID': 'Team2','LScore': '2Score','WTeamSeed': 'Team1Seed','LTeamSeed':'Team2Seed','WLoc':'Team1Home'}
    d2 = {'WTeamID': 'Team2','WScore': '2Score','LTeamID': 'Team1','LScore': '1Score','WTeamSeed': 'Team2Seed','LTeamSeed':'Team1Seed','WLoc':'Team1Home'}
    
    df = pd.concat([half1.rename(columns = d1),half2.rename(columns = d2)])
    
    df['score_diff'] = df['1Score']- df['2Score']
    
    drop_columns = ['1Score','2Score']
    df.drop(columns= drop_columns,inplace = True)

    return df
```

```python
def tourney_compact_preprocessing(dataframe):
    df = dataframe.drop(columns = ['NumOT','Season','DayNum'])

    half1 = df.sample(n = int(len(df)*0.5-0.5))
    half2 = df[~df.isin(half1)].dropna()
    
    half1['WLoc'] = half1['WLoc'].replace({'A' : -1,'H': 1,'N':0})
    half2['WLoc'] = half2['WLoc'].replace({'A' : -1,'H': 0,'N':1})
    
    d1 = {'WTeamID': 'Team1','WScore': '1Score','LTeamID': 'Team2','LScore': '2Score','WTeamSeed': 'Team1Seed','LTeamSeed':'Team2Seed','WLoc':'Team1Home'}
    d2 = {'WTeamID': 'Team2','WScore': '2Score','LTeamID': 'Team1','LScore': '1Score','WTeamSeed': 'Team2Seed','LTeamSeed':'Team1Seed','WLoc':'Team1Home'}
    df = pd.concat([half1.rename(columns = d1),half2.rename(columns = d2)])
 
    df['score_diff'] = df['1Score']- df['2Score']
    df['seed_diff'] = pd.to_numeric(df['Team1Seed'])- pd.to_numeric(df['Team2Seed'])
    
    drop_columns = ['1Score','2Score','Team1Seed','Team2Seed']
    df.drop(columns= drop_columns,inplace = True)

    return df
```
